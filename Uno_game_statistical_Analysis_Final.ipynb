```python
### Analysis of Data
colors = ['tab:purple','tab:red','tab:blue', 'tab:green', 'gold']

# Splitting df into different tables
hyp_cols = {}
for i in range(1,7):
    hyp_cols[f'Hyp{i}_cols'] = df.loc[df['group'] == i]
    

## FOR 2 PlAYER UNO GAME WITHOUT SPECIAL CARDS

#Null Hypothesis: color or number strategy does not effect(increase or decrease) p1 chances of winning.

# Alternative: Color or number strategy increases or decreases p1 chances of winning.

# Null case used for testing:(random, random) strategy for 2 players w/o special cards


#double-sided z-test for proportions

## Calculating proportion of p1 wins given random strategy
p1_random = hyp_cols["Hyp1_cols"].iloc[0]["p0_wins"]


## Null: mean proportion is .4978
prop_random = p1_random / n

# #Calculating proportion of p1 wins given color strategy
p1_color = hyp_cols["Hyp1_cols"].iloc[1]["p0_wins"]
prop_color = p1_color / n


print("Proportion z-test for color at alpha = .05: ")
zscore_color, pvalue_color = proportions_ztest(count=p1_color, nobs=n, value=prop_random)
print("z-value = " + str(zscore_color) + " and" +" p-value:" + str(pvalue_color))


## Calculating proportion of p1 wins given number strategy
# one sided z test as prop is smaller
#Null: number strategy has no effect on p1 wins
#Alternative: number strategy decreases p1 wins

p1_number = hyp_cols["Hyp1_cols"].iloc[2]["p0_wins"]
prop_number = p1_number / n

print("Proportion z-test for number at alpha = .05(one-sided): ")
zscore_number, pvalue_number = proportions_ztest(count=p1_number, nobs=n, value=prop_random, alternative = 'smaller')
print("z-value = " + str(zscore_number) + " and" " p-value:" + str(pvalue_number))


## Results: both color and number strategy is significant as pvalue < .05, hence, color or number strategy
## effect chances of winning.
## Moreover, the z-score for number strategy was negative hence, number strategy decreases chances of winning for p1. 

print("p-value for color strategy: " + str(pvalue_color) + " and p-value for number strategy: " + str(pvalue_number))
print("Both values are <.05 hence, results are significant, we reject null and conclude alternative.")


## Bar graph
Hyp1_prop = [prop_random,prop_color,prop_number]

props1_label = ['Random prop', 'Color prop(*)', 'Number prop(*)']

plt.bar(props1_label, Hyp1_prop, color = ['tab:red','tab:blue', 'tab:green'])


plt.xlabel('Strategies for 2 player Uno game without special cards')
plt.ylabel('Proportions')
plt.title('Proportions for 2 player Uno game without special cards')
plt.axhline(y=.5, color='black', linestyle='--')


for i, value in enumerate(Hyp1_prop):
    plt.text(i, value + 0.005, f"{value:.4f}", ha='center')


plt.show()


p1 = [0,pvalue_color,pvalue_number]
h1=[2,2,2]



 
```
